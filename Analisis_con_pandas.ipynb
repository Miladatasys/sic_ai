**Carga de datos dataset data_studentlist**

```python
from google.colab import files
uploaded = files.upload()

import pandas as pd

# Cargar el archivo CSV en un DataFrame
df = pd.read_csv('data_studentlist.csv')
```

**Exploración inicial**

```python
# Ver las primeras filas del DataFrame
print(df.head())

# Ver información general del DataFrame
print(df.info())

# Descripción estadística de los datos numéricos
print(df.describe())

# Verificar la presencia de valores nulos
print(df.isnull().sum())
```

**Organización de datos**

1. Distribución de Edades


```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.histplot(df['age'], bins=4, kde=False)
plt.title('Distribución de Edades de los Estudiantes')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.show()
```

2. Distribución de Notas


```python
sns.histplot(df['grade'], bins=4, kde=False)
plt.title('Distribución de Notas de los Estudiantes')
plt.xlabel('Grado')
plt.ylabel('Frecuencia')
plt.show()
```

3. Distribución por Género


```python
sns.countplot(x='gender', data=df)
plt.title('Número de Estudiantes por Género')
plt.xlabel('Género')
plt.ylabel('Número de Estudiantes')
plt.show()
```

4. Relación entre Altura y Peso:


```python
sns.scatterplot(x='height', y='weight', data=df)
plt.title('Relación entre Altura y Peso de los Estudiantes')
plt.xlabel('Altura (cm)')
plt.ylabel('Peso (kg)')
plt.show()
```

5. Ausencias


```python
sns.countplot(x='absence', data=df)
plt.title('Número de Ausencias')
plt.xlabel('Ausencia (Y/N)')
plt.ylabel('Número de Estudiantes')
plt.show()
```
