```python
def consultar_saldo(saldo):
  print(f"Su saldo actual es de ${saldo} ¡Gracias por consultar!")


def retirar_dinero(saldo):
  cantidad = float(input(f"Ingrese la cantidad a retirar: "))
  if cantidad <= saldo:
    saldo -= cantidad
    print(f"Su saldo actual es de $(saldo)")
  else:
    print("No tiene saldo suficiente")
  return saldo


def depositar_dinero(saldo):
  cantidad = float(input("Ingrese la cantidad a depositar: "))
  saldo += cantidad
  print(f"Deposito exitoso, ha depositado ${cantidad}")
  return saldo


def transferir_dinero(saldo):
  cuenta_destino = float(input("Ingrese numero de cuenta: "))
  cantidad = float(input("Ingrese monto a transferir: "))
  if cantidad <= saldo:
    saldo -= cantidad
    print(f"Transferencia exitosa a cuenta {cuenta_destino} con un monto de ${cantidad}")
  else:
    print("No tiene saldo suficiente")
    return saldo


def cajero_automatico():
  saldo = 80000
  while True:
    print("Bienvenidos al simulador de cajero automático\n")
    cuenta = input("Introduzca su número de cuenta para continuar: ")

    print("Estimado estudiante, este cajero automático es capaz de reliazar diferentes operaciones. Por favor, seleccione la opción deseada")

    print("1. Consultar saldo")
    print("2. Retirar dinero")
    print("3. Depositar dinero")
    print("4. Transferir dinero")
    opcion = int(input("Ingrese la opción deseada: "))

    if opcion == 1:
      consultar_saldo(saldo)
    elif opcion == 2:
      saldo = retirar_dinero(saldo)
    elif opcion == 3:
      saldo = depositar_dinero(saldo)
    elif opcion == 4:
      saldo = transferir_dinero(saldo)

    else:
      print("Opción no válida, intente nuevamente")

    continuar = input("¿Desea realizar otra operación? (s/n): ")
    if continuar.lower() != "s":
      print("Gracias por utilizar el cajero automático")
      break


cajero_automatico()
```
